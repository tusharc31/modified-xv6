```python
mp = dict(list())
yp = dict(list())
fi = open("data.txt", "r")
cnt=0
for line in fi:
    if line.startswith("P"):
        cnt=cnt+1
        continue
    #print(line)
    xl = [int(x) for x in line.split()]
    #print(xl)
    if(len(xl)==0):
        print("Empty line")
        continue
    if xl[0] not in mp:
        mp[xl[0]]=list()
        yp[xl[0]]=list()
    mp[xl[0]].append(xl[6])
    yp[xl[0]].append(cnt)
    
```

```python
print(mp.keys())
print(len(mp.keys()))
```

```python
import matplotlib.pyplot as plt
print(mp[1])
```

```python
colors = ["white", "black", "silver", "rosybrown", "brown", "red", "salmon", "yellowgreen", "green", "darkgreen",
          "aquamarine", "darkslategrey", "darkslateblue", "gold", "bisque"]
```

```python
print(len(colors))
```

```python
lx=list()
plt.figure(figsize=(10,20))
for keys in yp:
    plt.plot(yp[keys],mp[keys],colors[keys])
    lx.append(f"P{keys}")
plt.legend(lx, loc="upper right")
plt.show()
plt.savefig("data.png")
```
